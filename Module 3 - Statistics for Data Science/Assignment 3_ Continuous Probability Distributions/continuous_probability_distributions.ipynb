###The lifetime (in hours) of a transistor is a random variable T with exponential distribution. The average lifetime of the transistor is 500 hours. What is the probability that the transistor will last between 300 and 1000 hours?

```
#P(300<= T <=1000) = P(T>=300) - P(T>1000)
#cdf accumulates up to the given point
expon.cdf(1000,scale=500)-expon.cdf(300,scale=500)
```

R = 0,4135

###Grades on a statistics test in a college class are normally distributed with a mean of 6.0 and a variance of 16.0. What is the probability that a student will get a grade between 7.0 and 8.0?

```
#P(7,0 <= X <= 8,0)
#Standard Deviation = square root of 16
norm.cdf(8,6,4)-norm.cdf(7,6,4)
```

###In 400 simulations of a certain equipment, 80 successes were obtained. What is the 95% confidence interval for the proportion of successes?

```
import statsmodels.stats.proportion as prop

inf, sup = prop.proportion_confint (80, 400, alpha = 0.05, method = 'normal')
print ('Limite inferior:', round(inf,2));
print ('Limite superior:', round(sup,2));
```

###A sample of 30 valves is taken from the total number of valves manufactured by a company, and the average life of 800h with a standard deviation of 100h is obtained. What is the 99% confidence interval for the average life of the population?

```
from scipy.stats import t,norm

inf, sup = norm.interval(0.99, loc = 800, scale = 100/np.sqrt(30))
print('Limite inferior:', round(inf,1))
print('Limite superior:', round(sup,1))
```
